```python
import numpy as np
from models.linear_regression import LinearRegression

X = np.array([[1, 2],
              [3, 4],
              [5, 6]])

y = np.array([1, 2, 3])

model = LinearRegression()
model.fit(X, y)

print("Coef:", model.coef_)
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("Predictions:", y_pred)
```
